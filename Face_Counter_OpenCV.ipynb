Human Face Counter

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/haarcascade_frontalface_default.xml')
img = cv2.imread("/content/drive/MyDrive/human pic.webp")
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2_imshow(gray)
```

```python
faces = face_cascade.detectMultiScale(gray) 
print(faces)
```

```python
for(x,y,w,h) in faces:
  cv2.rectangle(img,(x,y),(x+w, y+h),(100, 0, 255), 2)
# Display the output 
cv2_imshow(img)
```

```python
eye_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/haarcascade_eye.xml')
img = cv2.imread("/content/drive/MyDrive/human pic.webp")
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
eyes = eye_cascade.detectMultiScale(gray)
print(eyes)
for(x, y, w, h) in eyes:
  cv2.rectangle(img, (x,y), (x+w, y+h), (90, 20, 50), 2)
#Display the output
cv2_imshow(img) 
```

```python
def love_to_detect(img_path):
  face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/haarcascade_frontalface_default.xml')
  img = cv2.imread(img_path)
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  faces = face_cascade.detectMultiScale(gray)
  for (x, y, w, h) in faces:
     cv2.rectangle(img, (x,y), (x+w, y+h), (90, 200, 90), 2)
  #Display the output
  cv2_imshow(img)
```
